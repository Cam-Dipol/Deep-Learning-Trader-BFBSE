## Analysis of Results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

sessiontime_1sec = pd.read_csv('C:/Users/camer/Documents/Masters Thesis/Deep-Learning-Trader-BFBSE/00-00-00-00-00-01-19 1sec.csv', header=None)
sessiontime_2sec = pd.read_csv('C:/Users/camer/Documents/Masters Thesis/Deep-Learning-Trader-BFBSE/00-00-00-00-00-01-19 2 sec.csv', header=None)
sessiontime_10sec = pd.read_csv('C:/Users/camer/Documents/Masters Thesis/Deep-Learning-Trader-BFBSE/00-00-00-00-00-01-19 10sec.csv', header=None)

column_names = [
    'Trial_Number',
    'Trader_Type_1', 'Total_Profit_1', 'Num_Traders_1', 'Profit_per_trader_1', 'Trades_Per_Trader_1', 'Time_1_1', 'Time_2_1',
    'Trader_Type_2', 'Total_Profit_2', 'Num_Traders_2', 'Profit_per_trader_2', 'Trades_Per_Trader_2', 'Time_1_2', 'Time_2_2'
]

sessiontime_1sec.columns = column_names
sessiontime_2sec.columns = column_names
sessiontime_10sec.columns = column_names


sessiontime_10sec.head()
```

```python
sessiontime_1sec['Higher_Profit_Per_Trader'] = sessiontime_1sec.apply(
    lambda row: row['Trader_Type_1'] if row['Profit_per_trader_1'] > row['Profit_per_trader_2'] else row['Trader_Type_2'], axis=1
)

trader_1_wins_1sec = (sessiontime_1sec['Higher_Profit_Per_Trader'] == sessiontime_1sec['Trader_Type_1']).sum()
trader_2_wins_1sec = (sessiontime_1sec['Higher_Profit_Per_Trader'] == sessiontime_1sec['Trader_Type_2']).sum()

sessiontime_2sec['Higher_Profit_Per_Trader'] = sessiontime_2sec.apply(
    lambda row: row['Trader_Type_1'] if row['Profit_per_trader_1'] > row['Profit_per_trader_2'] else row['Trader_Type_2'], axis=1
)

trader_1_wins_2sec = (sessiontime_2sec['Higher_Profit_Per_Trader'] == sessiontime_2sec['Trader_Type_1']).sum()
trader_2_wins_2sec = (sessiontime_2sec['Higher_Profit_Per_Trader'] == sessiontime_2sec['Trader_Type_2']).sum()

sessiontime_10sec['Higher_Profit_Per_Trader'] = sessiontime_10sec.apply(
    lambda row: row['Trader_Type_1'] if row['Profit_per_trader_1'] > row['Profit_per_trader_2'] else row['Trader_Type_2'], axis=1
)

trader_1_wins_10sec = (sessiontime_10sec['Higher_Profit_Per_Trader'] == sessiontime_10sec['Trader_Type_1']).sum()
trader_2_wins_10sec = (sessiontime_10sec['Higher_Profit_Per_Trader'] == sessiontime_10sec['Trader_Type_2']).sum()


print(f"Trader Type 1 ('{sessiontime_1sec['Trader_Type_1'].iloc[0]}') won {trader_1_wins_1sec} trials for 1sec.")
print(f"Trader Type 2 ('{sessiontime_1sec['Trader_Type_2'].iloc[0]}') won {trader_2_wins_1sec} trials for 1sec.")
print(f"Trader Type 1 ('{sessiontime_2sec['Trader_Type_1'].iloc[0]}') won {trader_1_wins_2sec} trials for 2sec.")
print(f"Trader Type 2 ('{sessiontime_2sec['Trader_Type_2'].iloc[0]}') won {trader_2_wins_2sec} trials for 2sec.")
print(f"Trader Type 1 ('{sessiontime_10sec['Trader_Type_1'].iloc[0]}') won {trader_1_wins_10sec} trials for 10sec.")
print(f"Trader Type 2 ('{sessiontime_10sec['Trader_Type_2'].iloc[0]}') won {trader_2_wins_10sec} trials for 10sec.")
```
